```python
from contextlib import nullcontext                                      # Used as an empty context when mixed precision is disabled.
import json                                                                 # Used to read tokenizer fallback metadata.
from pathlib import Path                                                     # Robust file handling for datasets, artifacts and output files.
import importlib.util                                                        # Used to detect optional fine-tuning dependencies early.
import math                                                                  # Used for batch counting and progress reporting.
import re                                                                    # Used to normalize prompts, questions and answers.
import time                                                                  # Used for retry pauses and runtime tracking.

import pandas as pd                                                          # Data loading, cleaning, checkpoint handling and CSV export.
import torch                                                                 # Local training and inference runtime.
from IPython.display import display                                          # Notebook previews.
from transformers import AutoModelForCausalLM, AutoTokenizer                 # Hugging Face model and tokenizer loading.


model_name = "ipst/Qwen2.5-0.5B-Instruct-SLDS"                              # Reuse the same legal-oriented instruct model as in Model 1.
training_data_path = Path("training_data") / "train_tax_qa.csv"             # Generated supervised training data with complex Austrian tax QA cases.
test_data_path = Path("dataset_clean.csv")                                  # Competition test questions used only for generation.
artifact_dir = Path("model_2_artifacts")                                    # Store the trained adapter and tokenizer metadata here.
output_dir = Path("model_2_outputs")                                        # Keep all generated inference files in one dedicated folder.
submission_path = output_dir / "submission.csv"                             # Final competition file with id and answer.
checkpoint_path = output_dir / "submission_checkpoint.csv"                   # Resume file for long-running inference jobs.

use_dev_sample = False                                                       # Switch to True for a smaller end-to-end development run.
dev_train_size = 64                                                          # Number of training rows kept in development mode.
dev_test_size = 25                                                           # Number of inference rows kept in development mode.
validation_fraction = 0.1                                                    # Hold out a small validation subset when enough training rows are available.

batch_size = 8                                                               # Number of prompts processed together during inference.
max_input_tokens = 256                                                       # Prompt truncation limit for inference.
max_new_tokens = 140                                                         # Answer generation limit aligned with Model 1.
temperature = 0.0                                                            # Deterministic decoding by default.
top_p = 1.0                                                                  # Only used when sampling is enabled.
num_retries = 1                                                              # Retry failed inference batches once.
retry_sleep_seconds = 1.5                                                    # Small pause between retries.
save_every_batches = 2                                                       # Save checkpoint and submission regularly.
fallback_answer = "Antwort konnte aufgrund eines technischen Fehlers nicht erzeugt werden."   # Controlled fallback for failed inference.

max_train_prompt_tokens = 256                                                # Prompt truncation limit for supervised fine-tuning examples.
max_train_total_tokens = 384                                                 # Total sequence length after appending the target answer.
lora_rank = 8                                                                # Low-rank adapter size for a lightweight local fine-tuning setup.
lora_alpha = 16                                                              # Scaling factor for the LoRA adapters.
lora_dropout = 0.05                                                          # Small dropout to regularize adapter training.
train_epochs = 2                                                             # Keep the default training budget compact enough for notebook usage.
train_learning_rate = 2e-4                                                   # Typical LoRA learning rate for small instruct models.
train_batch_size = 2                                                         # Default micro-batch size for LoRA fine-tuning.
eval_batch_size = 2                                                          # Evaluation batch size mirrors the small training setup.
gradient_accumulation_steps = 8                                              # Increase effective batch size without exhausting GPU memory.
fallback_train_batch_size = 1                                                # Use a smaller batch size when PEFT is unavailable and the full model is trained directly.
fallback_eval_batch_size = 1                                                 # Keep evaluation conservative in the direct fine-tuning fallback.
fallback_gradient_accumulation_steps = 16                                    # Compensate for the smaller fallback micro-batch size.
weight_decay = 0.01                                                          # Mild regularization for manual fine-tuning.
logging_steps = 10                                                           # Print compact training progress in the notebook.
save_strategy = "epoch"                                                    # Save checkpoints at the end of each epoch.
require_gpu_for_training = True                                              # Fine-tuning is intended for a local GPU environment.
random_seed = 42                                                             # Deterministic split and training seed.
dataloader_num_workers = 0                                                   # Keep notebook execution predictable across Windows setups.

required_test_columns = {"id", "prompt"}                                  # Required schema of the competition test CSV.
required_training_columns = {"question", "answer"}                        # Required minimum schema of the generated supervised training CSV.
optional_training_columns = ["focus_area", "legal_basis", "source"]      # Extra metadata columns provided by the generated training dataset.
checkpoint_columns = ["id", "answer", "status", "error_message", "attempts"]   # Standardized schema for resumable inference outputs.

artifact_dir.mkdir(parents=True, exist_ok=True)                              # Ensure the adapter directory exists before saving artifacts.
output_dir.mkdir(parents=True, exist_ok=True)                                # Ensure the output directory exists before writing CSV files.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")      # Prefer GPU when available.
pin_memory = device.type == "cuda"                                          # Enable faster host-to-device transfer when a GPU is available.
use_peft = importlib.util.find_spec("peft") is not None                     # Use LoRA when PEFT is available, otherwise fall back to direct fine-tuning.

print(f"model_name: {model_name}")
print(f"device: {device}")
print(f"training_data_path: {training_data_path.resolve()}")
print(f"test_data_path: {test_data_path.resolve()}")
print(f"artifact_dir: {artifact_dir.resolve()}")
print(f"submission_path: {submission_path.resolve()}")
print(f"checkpoint_path: {checkpoint_path.resolve()}")

print(f"use_dev_sample: {use_dev_sample}")
print(f"use_peft: {use_peft}")
if not use_peft:
    print("PEFT is not installed. The notebook will continue with direct fine-tuning without LoRA.")


# Helper for tokenizer metadata that may store tokens as small dictionaries.
def extract_special_token_value(token_entry):
    if isinstance(token_entry, dict):
        return token_entry.get("content")
    return token_entry


# Helper that finds a downloaded model snapshot before trying online loading.
def resolve_cached_model_snapshot(model_id):
    cache_root = Path.home() / ".cache" / "huggingface" / "hub" / f"models--{model_id.replace('/', '--')}"
    snapshots_dir = cache_root / "snapshots"
    if not snapshots_dir.exists():
        return None

    snapshot_paths = sorted((path for path in snapshots_dir.iterdir() if path.is_dir()), key=lambda item: item.stat().st_mtime, reverse=True)
    return snapshot_paths[0] if snapshot_paths else None


# Helper that keeps tokenizer loading working when the normal loader needs protobuf.
def load_tokenizer_with_fallback(model_id):
    model_source = resolved_model_source if resolved_model_source is not None else model_id
    try:
        return AutoTokenizer.from_pretrained(str(model_source), use_fast=True, local_files_only=resolved_model_source is not None)
    except ImportError as exc:
        if "protobuf" not in str(exc).lower():
            raise

        snapshot_path = resolved_model_source if resolved_model_source is not None else resolve_cached_model_snapshot(model_id)
        if snapshot_path is None:
            raise RuntimeError("Tokenizer loading requires protobuf, but no cached model snapshot was found for the local fallback.") from exc

        tokenizer_file = snapshot_path / "tokenizer.json"
        if not tokenizer_file.exists():
            raise RuntimeError(f"Tokenizer fallback file is missing: {tokenizer_file}") from exc

        tokenizer = PreTrainedTokenizerFast(tokenizer_file=str(tokenizer_file))
        tokenizer_config_path = snapshot_path / "tokenizer_config.json"
        special_tokens_map_path = snapshot_path / "special_tokens_map.json"

        tokenizer_config = {}
        if tokenizer_config_path.exists():
            tokenizer_config = json.loads(tokenizer_config_path.read_text(encoding="utf-8"))
            tokenizer.model_max_length = tokenizer_config.get("model_max_length", tokenizer.model_max_length)
            tokenizer.padding_side = tokenizer_config.get("padding_side", tokenizer.padding_side)
            tokenizer.truncation_side = tokenizer_config.get("truncation_side", tokenizer.truncation_side)
            chat_template = tokenizer_config.get("chat_template")
            if chat_template:
                tokenizer.chat_template = chat_template

        special_tokens_map = {}
        if special_tokens_map_path.exists():
            special_tokens_map = json.loads(special_tokens_map_path.read_text(encoding="utf-8"))

        for token_name in ["bos_token", "eos_token", "pad_token", "unk_token", "sep_token", "cls_token", "mask_token"]:
            token_value = special_tokens_map.get(token_name, tokenizer_config.get(token_name))
            token_value = extract_special_token_value(token_value)
            if token_value:
                setattr(tokenizer, token_name, token_value)

        additional_special_tokens = special_tokens_map.get("additional_special_tokens") or tokenizer_config.get("additional_special_tokens")
        if additional_special_tokens:
            tokenizer.additional_special_tokens = [extract_special_token_value(token) for token in additional_special_tokens]

        return tokenizer

resolved_model_source = resolve_cached_model_snapshot(model_name)                       # Prefer an existing local cache to avoid network dependence.
print(f"resolved_model_source: {resolved_model_source}")
```

# 2. Load training and test data

This stage loads the supervised training data and the test questions. It checks required columns, keeps optional metadata columns when they exist, removes empty values, and preserves the original test order for the final submission.

Function definitions:
- `normalize_whitespace(value)`: Converts a value to text and removes repeated spaces or line breaks.
- `normalize_columns(dataframe, columns)`: Applies the same text cleanup to selected columns when they exist.
- `require_non_empty_values(dataframe, columns, dataset_name)`: Stops the notebook if important columns contain empty values after cleaning.
- `load_csv_with_required_columns(csv_path, required_columns, dtype_map, optional_columns=None)`: Loads a CSV file, checks its required columns, and returns only the useful columns.


```python
# Shared cleaner used for ids, questions, answers and metadata fields.
def normalize_whitespace(value):
    value = "" if value is None else str(value)                               # Convert all values into strings before cleaning.
    return re.sub(r"\s+", " ", value).strip()                              # Collapse repeated whitespace into a single space.


# Small helper so the same cleanup is not repeated for every DataFrame.
def normalize_columns(dataframe, columns):
    for column in columns:
        if column in dataframe.columns:
            dataframe[column] = dataframe[column].map(normalize_whitespace)    # Normalize only the columns that are actually present.


# Validation helper that stops if important text fields are empty.
def require_non_empty_values(dataframe, columns, dataset_name):
    for column in columns:
        if dataframe[column].eq("").any():
            raise ValueError(f"Found empty {column} values in the {dataset_name} after normalization.")


# CSV loader that checks the expected schema before the pipeline continues.
def load_csv_with_required_columns(csv_path, required_columns, dtype_map, optional_columns=None):
    if not csv_path.exists():
        raise FileNotFoundError(f"Dataset file not found: {csv_path.resolve()}")

    try:
        loaded_df = pd.read_csv(csv_path, dtype=dtype_map, low_memory=False)
    except pd.errors.EmptyDataError as exc:
        raise ValueError(f"The input CSV is empty: {csv_path}") from exc
    except pd.errors.ParserError as exc:
        raise ValueError(f"The input CSV could not be parsed: {exc}") from exc

    missing_columns = set(required_columns) - set(loaded_df.columns)
    if missing_columns:
        raise ValueError(f"Missing required columns in {csv_path.name}: {sorted(missing_columns)}")

    selected_columns = list(required_columns)
    if optional_columns:
        selected_columns.extend(column for column in optional_columns if column in loaded_df.columns)
    return loaded_df.loc[:, selected_columns].copy()


train_df = load_csv_with_required_columns(
    training_data_path,
    required_training_columns,
    dtype_map={"question": "string", "answer": "string", "focus_area": "string", "legal_basis": "string", "source": "string"},
    optional_columns=optional_training_columns,
)
test_df = load_csv_with_required_columns(
    test_data_path,
    required_test_columns,
    dtype_map={"id": "string", "prompt": "string"},
)

normalize_columns(train_df, ["question", "answer", *optional_training_columns])
normalize_columns(test_df, ["id", "prompt"])
require_non_empty_values(train_df, ["question", "answer"], "training data")
require_non_empty_values(test_df, ["id", "prompt"], "test data")

if train_df.duplicated(subset=["question", "answer"]).any():
    duplicated_pairs = int(train_df.duplicated(subset=["question", "answer"]).sum())
    print(f"Warning: found {duplicated_pairs} duplicated training rows; duplicates are dropped.")
    train_df = train_df.drop_duplicates(subset=["question", "answer"], keep="first")
if test_df["id"].duplicated().any():
    duplicate_ids = test_df.loc[test_df["id"].duplicated(), "id"].tolist()
    raise ValueError(f"Found duplicate ids in the test data: {duplicate_ids[:10]}")

train_df = train_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)
test_df["input_order"] = test_df.index                                        # Preserve original order for the final submission file.

if use_dev_sample:
    train_df = train_df.head(dev_train_size).copy()
    test_df = test_df.head(dev_test_size).copy()

print(f"training rows loaded: {len(train_df)}")
print(f"test rows loaded: {len(test_df)}")
available_metadata_columns = [column for column in optional_training_columns if column in train_df.columns]
if available_metadata_columns:
    print("training metadata columns:", available_metadata_columns)
    display(train_df.loc[:, ["question", "answer", *available_metadata_columns]].head())
    if "focus_area" in train_df.columns:
        display(train_df["focus_area"].value_counts().sort_index())
else:
    display(train_df.head())
display(test_df.head())
```

# 3. Prepare supervised fine-tuning dataset

This stage builds the prompt format used for training and inference. It creates full training texts from question-answer pairs, then shuffles and splits them into training and validation examples.

Function definitions:
- `build_prompt(question)`: Cleans one question and places it into the shared system/question/answer prompt format.
- `build_training_examples(dataframe)`: Builds `prompt_text`, `answer_text`, and `full_text` columns for supervised fine-tuning.


```python
system_prompt = (
    "Du bist ein hilfreicher Assistent fuer Fragen zum oesterreichischen Steuerrecht. "
    "Beantworte die Frage auf Deutsch, praezise und sachlich. "
    "Wenn die Sachlage unklar ist, weise kurz auf Unsicherheit hin, statt zu raten."
)                                                                               # Reuse one stable instruction block for training and inference.


# Builds the same instruction format for both training and later inference.
def build_prompt(question):
    cleaned_question = normalize_whitespace(question)
    if cleaned_question == "":
        raise ValueError("Question is empty after whitespace normalization.")
    return f"System: {system_prompt}\nBenutzerfrage: {cleaned_question}\nAntwort:"


# Turns the cleaned QA table into the text columns needed for supervised training.
def build_training_examples(dataframe):
    training_examples = dataframe.copy()
    training_examples["prompt_text"] = training_examples["question"].map(build_prompt)
    training_examples["answer_text"] = training_examples["answer"].map(normalize_whitespace)
    if training_examples["answer_text"].eq("").any():
        raise ValueError("Answer is empty after whitespace normalization.")

    for column in optional_training_columns:
        if column not in training_examples.columns:
            training_examples[column] = ""
        training_examples[column] = training_examples[column].map(normalize_whitespace)

    training_examples["full_text"] = training_examples["prompt_text"] + " " + training_examples["answer_text"]
    ordered_columns = ["prompt_text", "answer_text", "full_text", *optional_training_columns]
    return training_examples.loc[:, ordered_columns].copy()


training_examples_df = build_training_examples(train_df)
if len(training_examples_df) < 2:
    raise ValueError("At least two training examples are required for a meaningful fine-tuning split.")

validation_size = int(round(len(training_examples_df) * validation_fraction))   # Use the configured fraction to decide the validation size.
validation_size = max(1, validation_size) if len(training_examples_df) >= 10 else 0   # Need at least 10 rows before taking a validation row.
validation_size = min(validation_size, len(training_examples_df) - 1) if validation_size > 0 else 0

shuffled_examples_df = training_examples_df.sample(frac=1.0, random_state=random_seed).reset_index(drop=True)   # frac=1.0 shuffles the full dataset, and the seed keeps it repeatable.
validation_examples_df = shuffled_examples_df.iloc[:validation_size].copy().reset_index(drop=True)
train_examples_df = shuffled_examples_df.iloc[validation_size:].copy().reset_index(drop=True) if validation_size > 0 else shuffled_examples_df.copy().reset_index(drop=True)

train_records = train_examples_df.to_dict(orient="records")
validation_records = validation_examples_df.to_dict(orient="records")

print(f"train examples: {len(train_examples_df)}")
print(f"validation examples: {len(validation_examples_df)}")
print(train_examples_df.iloc[0]["full_text"])
display(train_examples_df.head())
if "focus_area" in train_examples_df.columns:
    display(train_examples_df[["focus_area", "legal_basis"]].head())
```

# 4. Load tokenizer and base model

This stage loads the tokenizer and base model before training. It checks the GPU requirement, sets padding behavior, disables cache for training, and enables gradient checkpointing when a CUDA GPU is available.


```python
if require_gpu_for_training and device.type != "cuda":
    raise RuntimeError("Fine-tuning is configured for a local GPU. Disable 'require_gpu_for_training' only if you intentionally want a very slow CPU fallback.")

try:
    tokenizer = load_tokenizer_with_fallback(model_name)                             # Load the tokenizer, falling back to the cached tokenizer.json when protobuf is missing.
    model_source = str(resolved_model_source) if resolved_model_source is not None else model_name
    base_model = AutoModelForCausalLM.from_pretrained(model_source, local_files_only=resolved_model_source is not None)   # Load the causal language model from cache when available.
except Exception as exc:
    raise RuntimeError(f"Failed to load model or tokenizer '{model_name}': {exc}") from exc   # Surface model loading errors with context.

if tokenizer.pad_token is None and tokenizer.eos_token is not None:
    tokenizer.pad_token = tokenizer.eos_token                                      # Reuse EOS when the model does not define a pad token.
if tokenizer.pad_token is None:
    raise RuntimeError("Tokenizer does not define a usable pad_token or eos_token for training.")
if base_model.config.pad_token_id is None:
    base_model.config.pad_token_id = tokenizer.pad_token_id                        # Keep training and generation padding behavior explicit.
if hasattr(base_model.config, "use_cache"):
    base_model.config.use_cache = False                                            # Disable KV-cache during training to avoid incompatibilities with gradient checkpointing.
if device.type == "cuda" and hasattr(base_model, "gradient_checkpointing_enable"):
    base_model.gradient_checkpointing_enable()                                     # Reduce GPU memory pressure during training on longer examples.

base_model = base_model.to(device)                                                 # Move the base model to GPU or CPU once.
base_model.train()                                                                 # Keep the model in training mode before LoRA configuration.

print(tokenizer)
print(base_model.__class__.__name__)
```

# 5. Configure LoRA fine-tuning

This stage tokenizes the training examples and prepares the trainable model. If PEFT is available, it creates a LoRA adapter. If PEFT is missing, it falls back to training only a small part of the base model.

Function definitions:
- `tokenize_training_example(example)`: Converts one prompt-answer example into `input_ids`, `labels`, and `attention_mask` for supervised training.
- `supervised_data_collator(features)`: Pads examples in one batch to the same length and converts them into PyTorch tensors.


```python
if use_peft:
    from peft import LoraConfig, get_peft_model                                 # Import LoRA helpers only when PEFT is actually available.


# Converts one text example into token ids and training labels.
def tokenize_training_example(example):
    prompt_ids = tokenizer(
        example["prompt_text"],
        truncation=True,
        max_length=max_train_prompt_tokens,
        add_special_tokens=False,
    )["input_ids"]
    answer_ids = tokenizer(
        " " + example["answer_text"],
        truncation=True,
        max_length=max_train_total_tokens,
        add_special_tokens=False,
    )["input_ids"]

    eos_token_id = tokenizer.eos_token_id
    available_answer_length = max_train_total_tokens - len(prompt_ids) - (1 if eos_token_id is not None else 0)
    if available_answer_length <= 0:
        prompt_ids = prompt_ids[: max_train_total_tokens // 2]
        available_answer_length = max_train_total_tokens - len(prompt_ids) - (1 if eos_token_id is not None else 0)

    answer_ids = answer_ids[: max(1, available_answer_length)]
    input_ids = prompt_ids + answer_ids
    labels = ([-100] * len(prompt_ids)) + answer_ids.copy()                 # -100 masks the prompt tokens so only the answer trains the loss.
    if eos_token_id is not None and len(input_ids) < max_train_total_tokens:
        input_ids.append(eos_token_id)
        labels.append(eos_token_id)

    return {
        "input_ids": input_ids,
        "labels": labels,
        "attention_mask": [1] * len(input_ids),
    }


# Pads a small batch so all examples have the same tensor length.
def supervised_data_collator(features):
    batch_max_length = max(len(feature["input_ids"]) for feature in features)
    pad_token_id = tokenizer.pad_token_id
    return {
        "input_ids": torch.tensor([feature["input_ids"] + [pad_token_id] * (batch_max_length - len(feature["input_ids"])) for feature in features], dtype=torch.long),
        "attention_mask": torch.tensor([feature["attention_mask"] + [0] * (batch_max_length - len(feature["attention_mask"])) for feature in features], dtype=torch.long),
        "labels": torch.tensor([feature["labels"] + ([-100] * (batch_max_length - len(feature["labels"]))) for feature in features], dtype=torch.long),
    }


tokenized_train_dataset = [tokenize_training_example(record) for record in train_records]
tokenized_validation_dataset = [tokenize_training_example(record) for record in validation_records] if validation_records else []

if use_peft:
    lora_config = LoraConfig(
        r=lora_rank,
        lora_alpha=lora_alpha,
        lora_dropout=lora_dropout,
        bias="none",
        task_type="CAUSAL_LM",
        target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],        # LoRA updates the attention projection layers only.
    )
    training_model = get_peft_model(base_model, lora_config)
    training_model.print_trainable_parameters()
else:
    for parameter in base_model.parameters():
        parameter.requires_grad = False                                        # Avoid a full-model fallback update when PEFT is unavailable.

    fallback_trainable_modules = []
    if hasattr(base_model, "model") and hasattr(base_model.model, "layers") and len(base_model.model.layers) > 0:
        for parameter in base_model.model.layers[-1].parameters():
            parameter.requires_grad = True
        fallback_trainable_modules.append("last_decoder_layer")
    if hasattr(base_model, "model") and hasattr(base_model.model, "norm"):
        for parameter in base_model.model.norm.parameters():
            parameter.requires_grad = True
        fallback_trainable_modules.append("final_norm")
    if hasattr(base_model, "lm_head"):
        for parameter in base_model.lm_head.parameters():
            parameter.requires_grad = True
        fallback_trainable_modules.append("lm_head")

    training_model = base_model
    trainable_parameter_count = sum(parameter.numel() for parameter in training_model.parameters() if parameter.requires_grad)
    print(f"PEFT is not available. Using lightweight fallback fine-tuning on: {fallback_trainable_modules}")
    print(f"fallback_trainable_parameters: {trainable_parameter_count}")
```

# 6. Train and save adapter

This stage runs the manual PyTorch training loop. It builds data loaders, computes losses, accumulates gradients, evaluates after each epoch, and saves checkpoints plus the final trained artifacts.

Function definitions:
- `build_data_loader(dataset, batch_size_value, shuffle)`: Wraps tokenized records in a PyTorch `DataLoader`.
- `move_batch_to_device(batch)`: Moves all tensors in one batch to the selected CPU or GPU device.
- `compute_loss(model, batch)`: Runs the model on a batch and returns the training loss.
- `evaluate_model(model, data_loader)`: Computes the average validation loss without updating model weights.


```python
trainer_output_dir = artifact_dir / "trainer_state"                           # Keep transient epoch checkpoints separate from the final adapter export.
trainer_output_dir.mkdir(parents=True, exist_ok=True)
effective_train_batch_size = train_batch_size if use_peft else fallback_train_batch_size
effective_eval_batch_size = eval_batch_size if use_peft else fallback_eval_batch_size
effective_gradient_accumulation_steps = gradient_accumulation_steps if use_peft else fallback_gradient_accumulation_steps
trainable_parameters = [parameter for parameter in training_model.parameters() if parameter.requires_grad]
if not trainable_parameters:
    raise RuntimeError("No trainable parameters were found for section 6.")
train_parameter_dtype = trainable_parameters[0].dtype
use_autocast = device.type == "cuda" and train_parameter_dtype == torch.float16
autocast_context = (lambda: torch.autocast(device_type="cuda", dtype=torch.float16)) if use_autocast else nullcontext

print(f"effective_train_batch_size: {effective_train_batch_size}")
print(f"effective_eval_batch_size: {effective_eval_batch_size}")
print(f"effective_gradient_accumulation_steps: {effective_gradient_accumulation_steps}")


# Wraps tokenized records in the PyTorch batching tool.
def build_data_loader(dataset, batch_size_value, shuffle):
    return torch.utils.data.DataLoader(
        dataset,
        batch_size=batch_size_value,
        shuffle=shuffle,
        collate_fn=supervised_data_collator,
        num_workers=dataloader_num_workers,
        pin_memory=pin_memory,
    )


# Moves one batch to the selected runtime device.
def move_batch_to_device(batch):
    return {name: tensor.to(device, non_blocking=pin_memory) for name, tensor in batch.items()}


# Runs one forward pass and returns the loss value.
def compute_loss(model, batch):
    with autocast_context():
        outputs = model(**batch)
        loss = outputs.loss
    return loss


# Measures validation loss without changing model weights.
def evaluate_model(model, data_loader):
    if not data_loader:
        return None
    model.eval()
    total_loss = 0.0
    total_steps = 0
    with torch.inference_mode():
        for batch in data_loader:
            loss = compute_loss(model, move_batch_to_device(batch))
            total_loss += float(loss.detach().cpu())
            total_steps += 1
    model.train()
    return total_loss / max(1, total_steps)


train_loader = build_data_loader(tokenized_train_dataset, effective_train_batch_size, shuffle=True)
validation_loader = build_data_loader(tokenized_validation_dataset, effective_eval_batch_size, shuffle=False) if tokenized_validation_dataset else None
optimizer = torch.optim.AdamW(trainable_parameters, lr=train_learning_rate, weight_decay=weight_decay)
scaler = torch.amp.GradScaler("cuda", enabled=True) if use_autocast else None
training_history = []
training_start_time = time.time()
training_model.train()

try:
    for epoch_index in range(train_epochs):
        epoch_number = epoch_index + 1
        epoch_loss_total = 0.0
        optimizer_steps = 0
        optimizer.zero_grad(set_to_none=True)

        for step_index, batch in enumerate(train_loader, start=1):
            batch = move_batch_to_device(batch)
            raw_loss = compute_loss(training_model, batch)
            loss = raw_loss / effective_gradient_accumulation_steps                 # Scale the loss before accumulating gradients.
            if scaler is not None:
                scaler.scale(loss).backward()
            else:
                loss.backward()
            epoch_loss_total += float(raw_loss.detach().cpu())

            should_step = (step_index % effective_gradient_accumulation_steps == 0) or (step_index == len(train_loader))
            if should_step:
                if scaler is not None:
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    optimizer.step()
                optimizer.zero_grad(set_to_none=True)
                optimizer_steps += 1

            if logging_steps > 0 and (step_index % logging_steps == 0 or step_index == len(train_loader)):
                print(f"epoch {epoch_number}/{train_epochs} | train step {step_index}/{len(train_loader)} | loss: {float(raw_loss.detach().cpu()):.4f}")

        epoch_summary = {
            "epoch": epoch_number,
            "train_loss": epoch_loss_total / max(1, len(train_loader)),
            "eval_loss": evaluate_model(training_model, validation_loader),
            "optimizer_steps": optimizer_steps,
        }
        training_history.append(epoch_summary)
        print(f"finished epoch {epoch_number}: train_loss={epoch_summary['train_loss']:.4f}, eval_loss={epoch_summary['eval_loss']}")

        if save_strategy == "epoch":
            epoch_checkpoint_dir = trainer_output_dir / f"checkpoint-epoch-{epoch_number}"
            training_model.save_pretrained(str(epoch_checkpoint_dir))
            tokenizer.save_pretrained(str(epoch_checkpoint_dir))
except Exception as exc:
    failure_dir = trainer_output_dir / "failed_state"
    failure_dir.mkdir(parents=True, exist_ok=True)
    training_model.save_pretrained(str(failure_dir))
    tokenizer.save_pretrained(str(failure_dir))
    raise RuntimeError(f"Training failed in section 6: {exc}") from exc

training_model.save_pretrained(str(artifact_dir))                               # Persist the trained model or LoRA adapter into the final artifact directory.
tokenizer.save_pretrained(str(artifact_dir))                                    # Persist the tokenizer metadata needed for later adapter reloading.

train_result = {
    "epochs": train_epochs,
    "train_batches": len(train_loader),
    "validation_batches": len(validation_loader) if validation_loader is not None else 0,
    "runtime_seconds": round(time.time() - training_start_time, 2),
    "history": training_history,
}
print(train_result)
print(f"Saved adapter artifacts to: {artifact_dir.resolve()}")
```

# 7. Load fine-tuned model for inference

This stage reloads the tokenizer and trained artifacts for generation. If a LoRA adapter was saved, it loads the base model and applies the adapter. If a full fallback model was saved, it loads that model directly.


```python
adapter_config_path = artifact_dir / "adapter_config.json"             # PEFT writes this file when a LoRA adapter was saved.
full_model_config_path = artifact_dir / "config.json"                # The fallback full-model save writes this config file.
if not adapter_config_path.exists() and not full_model_config_path.exists():
    raise FileNotFoundError(f"No trained model artifacts found in {artifact_dir.resolve()}. Run section 6 first.")

try:
    inference_tokenizer = load_tokenizer_with_fallback(str(artifact_dir))
    if inference_tokenizer.pad_token is None and inference_tokenizer.eos_token is not None:
        inference_tokenizer.pad_token = inference_tokenizer.eos_token
    if inference_tokenizer.pad_token is None:
        raise RuntimeError("Tokenizer does not define a usable pad_token or eos_token for inference.")

    if adapter_config_path.exists():
        if not use_peft:
            raise ImportError("Adapter artifacts were found, but the 'peft' package is not installed.")
        from peft import PeftModel                                             # Needed only when loading a saved LoRA adapter.
        model_source = str(resolved_model_source) if resolved_model_source is not None else model_name
        inference_base_model = AutoModelForCausalLM.from_pretrained(model_source, local_files_only=resolved_model_source is not None)
        inference_base_model = inference_base_model.to(device)
        fine_tuned_model = PeftModel.from_pretrained(inference_base_model, str(artifact_dir))
    else:
        fine_tuned_model = AutoModelForCausalLM.from_pretrained(str(artifact_dir))
        fine_tuned_model = fine_tuned_model.to(device)
except Exception as exc:
    raise RuntimeError(f"Failed to load the fine-tuned inference model: {exc}") from exc

fine_tuned_model.eval()
print(inference_tokenizer)
print(fine_tuned_model.__class__.__name__)
```

# 8. Run batched inference with checkpointing

This stage reuses the inference pattern from Model 1. It loads any checkpoint, generates answers only for pending rows, saves progress regularly, and writes the final submission file.

Function definitions:
- `load_checkpoint_frame(path_to_checkpoint)`: Loads and validates a checkpoint file, or returns an empty checkpoint table when none exists.
- `save_results(current_results_df)`: Saves the current checkpoint and the ordered submission file.
- `generate_batch_answers(prompt_list)`: Tokenizes prompts, runs the fine-tuned model, removes prompt tokens, and returns cleaned generated answers.


```python
# Loads previous inference progress if a checkpoint already exists.
def load_checkpoint_frame(path_to_checkpoint):
    if not path_to_checkpoint.exists():
        return pd.DataFrame(columns=checkpoint_columns)

    try:
        checkpoint_df_local = pd.read_csv(path_to_checkpoint, dtype="string")
    except pd.errors.EmptyDataError:
        return pd.DataFrame(columns=checkpoint_columns)
    except pd.errors.ParserError as exc:
        raise ValueError(f"Checkpoint could not be parsed: {exc}") from exc

    missing_checkpoint_columns = set(checkpoint_columns) - set(checkpoint_df_local.columns)
    if missing_checkpoint_columns:
        raise ValueError(f"Checkpoint is missing columns: {sorted(missing_checkpoint_columns)}")

    checkpoint_df_local = checkpoint_df_local.loc[:, checkpoint_columns].copy()
    checkpoint_df_local["id"] = checkpoint_df_local["id"].map(normalize_whitespace)
    checkpoint_df_local["answer"] = checkpoint_df_local["answer"].map(normalize_whitespace)
    checkpoint_df_local["status"] = checkpoint_df_local["status"].fillna("pending")
    checkpoint_df_local["error_message"] = checkpoint_df_local["error_message"].map(normalize_whitespace)
    checkpoint_df_local["attempts"] = checkpoint_df_local["attempts"].fillna("0")
    return checkpoint_df_local.drop_duplicates(subset=["id"], keep="last")


# Saves both resumable progress and the final competition file.
def save_results(current_results_df):
    checkpoint_export_df = current_results_df.loc[:, checkpoint_columns].copy()
    checkpoint_export_df.to_csv(checkpoint_path, index=False)

    submission_export_df = current_results_df.sort_values("input_order").loc[:, ["id", "answer"]].copy()
    submission_export_df["answer"] = submission_export_df["answer"].fillna("").astype(str)
    submission_export_df.to_csv(submission_path, index=False)


# Generates answers from the fine-tuned model for one prompt batch.
def generate_batch_answers(prompt_list):
    inference_parameter_dtype = next(fine_tuned_model.parameters()).dtype
    inference_autocast_context = (lambda: torch.autocast(device_type="cuda", dtype=torch.float16)) if device.type == "cuda" and inference_parameter_dtype == torch.float16 else nullcontext
    encoded_input = inference_tokenizer(
        prompt_list,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=max_input_tokens,
    )
    encoded_input = {key: value.to(device, non_blocking=pin_memory) for key, value in encoded_input.items()}
    generation_kwargs = {
        "max_new_tokens": max_new_tokens,
        "pad_token_id": inference_tokenizer.pad_token_id,
        "eos_token_id": inference_tokenizer.eos_token_id,
        "do_sample": temperature > 0.0,
    }
    if temperature > 0.0:
        generation_kwargs["temperature"] = temperature
        generation_kwargs["top_p"] = top_p

    with torch.inference_mode(), inference_autocast_context():
        generated_tokens = fine_tuned_model.generate(**encoded_input, **generation_kwargs)

    prompt_lengths = encoded_input["attention_mask"].sum(dim=1).tolist()
    return [normalize_whitespace(inference_tokenizer.decode(output_tokens[int(prompt_length):], skip_special_tokens=True)) for output_tokens, prompt_length in zip(generated_tokens, prompt_lengths)]


checkpoint_df = load_checkpoint_frame(checkpoint_path)
results_df = test_df.merge(checkpoint_df, on="id", how="left")
results_df["answer"] = results_df["answer"].fillna("")
results_df["status"] = results_df["status"].fillna("pending")
results_df["error_message"] = results_df["error_message"].fillna("")
results_df["attempts"] = pd.to_numeric(results_df["attempts"], errors="coerce").fillna(0).astype(int)

completed_mask = results_df["status"].eq("success") & results_df["answer"].ne("")
pending_index_list = results_df.index[~completed_mask].tolist()
total_pending = len(pending_index_list)
total_batches = math.ceil(total_pending / batch_size) if total_pending > 0 else 0

print(f"completed_rows: {int(completed_mask.sum())}")
print(f"pending_rows: {total_pending}")
display(results_df.head())

if total_pending == 0:
    print("No pending rows found. Existing outputs are reused.")
else:
    start_time = time.time()

    for batch_number in range(total_batches):
        batch_start = batch_number * batch_size
        batch_end = min(batch_start + batch_size, total_pending)
        batch_indices = pending_index_list[batch_start:batch_end]
        prompt_list = results_df.loc[batch_indices, "prompt"].map(build_prompt).tolist()

        batch_answers = None
        last_error_message = ""
        for attempt in range(1, num_retries + 2):                              # +2 gives the first try plus the configured retry count.
            try:
                batch_answers = generate_batch_answers(prompt_list)
                if len(batch_answers) != len(batch_indices):
                    raise RuntimeError("Generated answer count does not match batch size.")

                for row_index, answer_text in zip(batch_indices, batch_answers):
                    results_df.at[row_index, "answer"] = answer_text or fallback_answer
                    results_df.at[row_index, "status"] = "success" if answer_text else "fallback"
                    results_df.at[row_index, "error_message"] = ""
                    results_df.at[row_index, "attempts"] = attempt
                break
            except Exception as exc:
                last_error_message = normalize_whitespace(str(exc))
                if attempt <= num_retries:
                    time.sleep(retry_sleep_seconds)

        if batch_answers is None:
            for row_index in batch_indices:
                results_df.at[row_index, "answer"] = fallback_answer
                results_df.at[row_index, "status"] = "error"
                results_df.at[row_index, "error_message"] = last_error_message or "Unknown inference error."
                results_df.at[row_index, "attempts"] = num_retries + 1

        if (batch_number + 1) % save_every_batches == 0 or (batch_number + 1) == total_batches:
            save_results(results_df)

        processed_rows = batch_end
        error_rows = int(results_df["status"].eq("error").sum())
        elapsed_seconds = time.time() - start_time
        print(f"processed {processed_rows}/{total_pending} pending rows | errors: {error_rows} | elapsed: {elapsed_seconds:.1f}s")

    save_results(results_df)
```
