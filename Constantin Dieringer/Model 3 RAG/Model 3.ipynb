# Local RAG Pipeline for `dataset_clean.csv`

This notebook builds a local Retrieval-Augmented Generation pipeline. It loads the questions, builds a small RIS-based retrieval corpus, retrieves relevant legal passages, and uses the local Hugging Face model to generate answers.

No API model or API key is used here. The retrieval source list is `Model_3/rag_corpus.csv`, and the generated output is written to `Model_3/model_3_outputs/submission.csv`.


# 1. Setup

This stage imports the libraries, defines paths, and keeps all runtime settings in one place.

Import explanation:
- `Path` is used for input, corpus, output, and checkpoint paths.
- `Request` and `urlopen` are used to download RIS pages from their source URLs.
- `html` is used to decode HTML entities from legal web pages.
- `math` is used for batch counting and retrieval scoring.
- `re` is used to clean text and tokenize retrieval terms.
- `time` is used for retry pauses and runtime tracking.
- `numpy` is imported for lightweight numeric retrieval work.
- `pandas` is used for CSV loading, DataFrame handling, checkpointing, and export.
- `torch` runs the local language model on CPU or GPU.
- `display` shows compact notebook previews.
- `AutoTokenizer` and `AutoModelForCausalLM` load the local Hugging Face tokenizer and model.


```python
from pathlib import Path                                                          # Robust file handling for dataset and output files.
from urllib.request import Request, urlopen                                       # Used for loading RIS source documents.
import html                                                                       # Used for decoding legal text from RIS pages.
import math                                                                       # Used for batch counting and progress reporting.
import re                                                                         # Used to normalize prompts, answers and legal text.
import time                                                                       # Used for runtime tracking and retry pauses.

import numpy as np                                                                # Used for lightweight retrieval scoring.
import pandas as pd                                                               # Data loading, checkpoint handling and CSV export.
import torch                                                                      # Local inference runtime.
from IPython.display import display                                               # Notebook previews.
from transformers import AutoModelForCausalLM, AutoTokenizer                      # Hugging Face model and tokenizer loading.

model_name = "ipst/Qwen2.5-0.5B-Instruct-SLDS"                                   # Same legal-oriented instruct model as in Model 1 and Model 2.

working_dir = Path.cwd()                                                          # Notebook may run from the project root or from Model_3.
base_dir = working_dir if working_dir.name == "Model_3" else working_dir / "Model_3"

input_path = base_dir / "dataset_clean.csv"                                      # Input CSV with question ids and prompts.
corpus_path = base_dir / "rag_corpus.csv"                                        # Local list of official RIS source URLs.
output_dir = base_dir / "model_3_outputs"                                        # Keep all generated files in one dedicated folder.
submission_path = output_dir / "submission.csv"                                  # Final competition file with id and answer.
checkpoint_path = output_dir / "submission_checkpoint.csv"                        # Resume file for long-running inference jobs.

use_dev_sample = False                                                            # Switch to True for a fast development run.
dev_sample_size = 25                                                              # Number of rows kept in development mode.
batch_size = 4                                                                    # RAG prompts are longer, so batches stay smaller than in Model 1.
max_input_tokens = 1024                                                           # Prompt truncation limit including retrieved context.
max_new_tokens = 100                                                              # Answer generation limit.
temperature = 0.0                                                                 # Deterministic decoding by default.
top_p = 1.0                                                                       # Only used when sampling is enabled.
num_retries = 1                                                                   # Retry failed batches once.
retry_sleep_seconds = 1.5                                                         # Small pause between retries.
save_every_batches = 2                                                            # Save checkpoint and submission regularly.
retrieval_top_k = 4                                                               # Number of legal passages passed into the final prompt.
request_timeout = 25                                                              # Timeout for RIS document downloads.
fallback_answer = "Antwort konnte aufgrund eines technischen Fehlers nicht erzeugt werden."   # Controlled fallback for failed inference.

required_input_columns = {"id", "prompt"}                                      # Required schema of the input CSV.
required_corpus_columns = {"passage_id", "title", "source_url"}                # Required schema of the RIS source list.
checkpoint_columns = ["id", "answer", "status", "error_message", "attempts"]   # Standardized schema for resumable outputs.

output_dir.mkdir(parents=True, exist_ok=True)                                    # Ensure output directory exists before writing files.
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")           # Prefer GPU when available.
pin_memory = device.type == "cuda"                                               # Enable faster host-to-device transfer when a GPU is available.

print(f"model_name: {model_name}")
print(f"device: {device}")
print(f"input_path: {input_path.resolve()}")
print(f"corpus_path: {corpus_path.resolve()}")
print(f"submission_path: {submission_path.resolve()}")
print(f"checkpoint_path: {checkpoint_path.resolve()}")
print(f"use_dev_sample: {use_dev_sample}")
```

# 2. Load input data

This stage loads `dataset_clean.csv`, checks the required `id` and `prompt` columns, cleans the text, removes empty prompts, and stores the original order for the final submission.

Function definition:
- `normalize_whitespace(value)`: Converts a value to text, handles missing values, and removes repeated whitespace.


```python
def normalize_whitespace(value):
    value = "" if value is None or pd.isna(value) else str(value)                 # Convert all values into strings before cleaning.
    return re.sub(r"\s+", " ", value).strip()                                  # Collapse repeated whitespace into a single space.


if not input_path.exists():
    raise FileNotFoundError(f"Input file not found: {input_path.resolve()}")

try:
    df = pd.read_csv(input_path, dtype="string")                                  # Keep ids stable and avoid numeric coercion.
except Exception as exc:
    raise RuntimeError(f"Could not read input file: {input_path.resolve()}") from exc

missing_input_columns = required_input_columns - set(df.columns)
if missing_input_columns:
    raise ValueError(f"Missing input columns: {sorted(missing_input_columns)}")

df = df.loc[:, ["id", "prompt"]].copy()
df["id"] = df["id"].map(normalize_whitespace)
df["prompt"] = df["prompt"].map(normalize_whitespace)
df = df[df["prompt"].str.len() > 0].reset_index(drop=True)
df["input_order"] = range(len(df))

if use_dev_sample:
    df = df.head(dev_sample_size).copy().reset_index(drop=True)                   # Keep development runs short and reproducible.
    df["input_order"] = range(len(df))

assert df["id"].is_unique, "Each input id must be unique."
print(f"Loaded {len(df)} questions from {input_path}.")
display(df.head(3))
```

# 3. Build retrieval corpus

This stage loads the RIS source list from `rag_corpus.csv`, downloads the legal text, removes HTML markup, and creates one retrieval table with passage text and source metadata.

Function definitions:
- `clean_ris_html(raw_html)`: Removes script/style blocks and HTML tags, then returns readable text.
- `fetch_ris_text(source_url)`: Downloads one RIS page and returns cleaned legal text, or an empty fallback if the download fails.
- `build_corpus_frame(path_to_corpus)`: Loads the corpus CSV, validates its columns, fetches each source, and returns a deduplicated passage DataFrame.


```python
def clean_ris_html(raw_html):
    text = re.sub(r"(?is)<script.*?</script>|<style.*?</style>", " ", raw_html)  # Remove non-content blocks first.
    text = re.sub(r"(?is)<br\s*/?>|</p>|</div>|</li>|</h\d>", "\n", text)        # Preserve rough paragraph boundaries.
    text = re.sub(r"(?is)<[^>]+>", " ", text)                                   # Strip remaining HTML tags.
    text = html.unescape(text)
    return normalize_whitespace(text)


def fetch_ris_text(source_url):
    try:
        request = Request(source_url, headers={"User-Agent": "Mozilla/5.0 local RAG notebook"})
        with urlopen(request, timeout=request_timeout) as response:
            raw_html = response.read().decode("utf-8", errors="replace")
        return clean_ris_html(raw_html)
    except Exception as exc:
        print(f"RIS download failed, using title-only fallback: {source_url} | {exc}")
        return ""


def build_corpus_frame(path_to_corpus):
    if not path_to_corpus.exists():
        raise FileNotFoundError(f"RAG corpus file not found: {path_to_corpus.resolve()}")

    try:
        source_df = pd.read_csv(path_to_corpus, dtype="string")
    except Exception as exc:
        raise RuntimeError(f"Could not read RAG corpus file: {path_to_corpus.resolve()}") from exc

    missing_corpus_columns = required_corpus_columns - set(source_df.columns)
    if missing_corpus_columns:
        raise ValueError(f"Missing RAG corpus columns: {sorted(missing_corpus_columns)}")

    records = []
    for _, row in source_df.iterrows():
        passage_id = normalize_whitespace(row.get("passage_id", ""))
        law = normalize_whitespace(row.get("law", ""))
        title = normalize_whitespace(row.get("title", ""))
        paragraph = normalize_whitespace(row.get("paragraph", ""))
        source_url = normalize_whitespace(row.get("source_url", ""))

        if not passage_id or not source_url:
            continue

        legal_text = fetch_ris_text(source_url)
        if not legal_text:
            legal_text = f"{title}. Official source: {source_url}"              # Keep retrieval usable when one RIS request fails.

        passage_text = "\n".join([
            f"Title: {title}",
            f"Law: {law}",
            f"Paragraph: {paragraph}",
            f"Text: {legal_text}",
            f"Source: {source_url}",
        ])
        records.append({"passage_id": passage_id, "text": passage_text, "source": source_url, "legal_basis": title})

    if not records:
        raise ValueError("The retrieval corpus is empty after loading RIS sources.")

    return pd.DataFrame(records).drop_duplicates(subset=["text"]).reset_index(drop=True)


corpus_df = build_corpus_frame(corpus_path)
print(f"Built retrieval corpus with {len(corpus_df)} passages from {corpus_path}.")
display(corpus_df[["passage_id", "legal_basis", "source"]].head())
```

# 4. Build retriever

This stage creates a simple local retriever based on token overlap. It does not need embeddings or an external API, which keeps the retrieval step easy to reproduce in the notebook.

Function definitions:
- `tokenize_for_retrieval(text)`: Turns text into lowercase retrieval tokens and keeps useful tokens only.
- `build_inverted_index(text_series)`: Builds token sets and a lookup index from token to corpus row.
- `retrieve_passages(question, top_k=retrieval_top_k)`: Scores matching passages for one question and returns the best passages.


```python
def tokenize_for_retrieval(text):
    tokens = re.findall(r"\w+|\u00a7", normalize_whitespace(text).lower())
    return {token for token in tokens if len(token) > 2 or token == "\u00a7"}


def build_inverted_index(text_series):
    token_sets = []
    index = {}

    for row_index, text in enumerate(text_series):
        token_set = tokenize_for_retrieval(text)
        token_sets.append(token_set)
        for token in token_set:
            index.setdefault(token, []).append(row_index)

    return token_sets, index


def retrieve_passages(question, top_k=retrieval_top_k):
    question_tokens = tokenize_for_retrieval(question)
    score_by_index = {}

    for token in question_tokens:
        for row_index in inverted_index.get(token, []):
            score_by_index[row_index] = score_by_index.get(row_index, 0) + 1

    if score_by_index:
        candidate_items = sorted(
            score_by_index.items(),
            key=lambda item: item[1] / math.sqrt(max(len(corpus_token_sets[item[0]]), 1)),
            reverse=True,
        )[:top_k]
        candidate_indices = [row_index for row_index, _ in candidate_items]
        candidate_scores = [score / math.sqrt(max(len(corpus_token_sets[row_index]), 1)) for row_index, score in candidate_items]
    else:
        candidate_indices = list(range(min(top_k, len(corpus_df))))                # Deterministic fallback if no lexical overlap is found.
        candidate_scores = [0.0] * len(candidate_indices)

    retrieved_df = corpus_df.iloc[candidate_indices].copy().reset_index(drop=True)
    retrieved_df["score"] = candidate_scores
    return retrieved_df


corpus_token_sets, inverted_index = build_inverted_index(corpus_df["text"].tolist())
sample_retrieval = retrieve_passages(df.loc[0, "prompt"])
print(f"Retrieved {len(sample_retrieval)} passages for one sample question.")
display(sample_retrieval[["passage_id", "score", "legal_basis"]])
```

# 5. Build RAG prompts

This stage combines each user question with the retrieved RIS context. The final prompt tells the model to use the context first and only state uncertainty when the context is not enough.

Function definitions:
- `format_context(retrieved_df, max_context_chars=5200)`: Turns retrieved rows into one context string while staying under a character limit.
- `build_prompt(question)`: Cleans the question, retrieves context, and builds the final RAG prompt.


```python
system_prompt = (
    "Du bist ein hilfreicher Assistent fuer Fragen zum oesterreichischen Steuerrecht. "
    "Beantworte die Frage auf Deutsch, praezise und sachlich. "
    "Nutze vorrangig den bereitgestellten RIS-Kontext. "
    "Wenn der Kontext nicht ausreicht, weise kurz auf Unsicherheit hin, statt zu raten."
)                                                                                 # Reuse one stable instruction block for all RAG prompts.


# Formats retrieved RIS passages while keeping the prompt below a fixed text budget.
def format_context(retrieved_df, max_context_chars=5200):
    context_blocks = []
    used_chars = 0                                                               # Counts characters so the context limit is respected.

    for _, row in retrieved_df.iterrows():
        block = f"[{row['passage_id']}] {row['legal_basis']}\n{row['text']}"
        if used_chars + len(block) > max_context_chars:
            break
        context_blocks.append(block)
        used_chars += len(block)

    return "\n\n".join(context_blocks)


def build_prompt(question):
    cleaned_question = normalize_whitespace(question)
    if cleaned_question == "":
        raise ValueError("Question is empty after whitespace normalization.")

    retrieved_df = retrieve_passages(cleaned_question)
    context = format_context(retrieved_df)
    prompt_text = (
        f"System: {system_prompt}\n"
        f"RIS-Kontext:\n{context}\n\n"
        f"Benutzerfrage: {cleaned_question}\n"
        "Antwort:"
    )
    return prompt_text


sample_prompt = build_prompt(df.loc[0, "prompt"])
print(sample_prompt[:1800])
```

# 6. Load local model

This stage loads the tokenizer and the local Hugging Face model. It also includes fallback logic for cached tokenizer files, then moves the model to the selected device and switches it to inference mode.

Function definitions:
- `extract_special_token_value(token_entry)`: Reads the real token text from tokenizer metadata.
- `resolve_cached_model_snapshot(model_id)`: Finds the newest local Hugging Face cache snapshot for the model.
- `load_tokenizer_with_fallback(model_id)`: Loads the tokenizer normally or from cached tokenizer files if needed.


```python
# Helper for tokenizer metadata that may store tokens as dictionaries.
def extract_special_token_value(token_entry):
    if isinstance(token_entry, dict):
        return token_entry.get("content")
    return token_entry


# Finds a local Hugging Face cache snapshot before downloading anything.
def resolve_cached_model_snapshot(model_id):
    cache_root = Path.home() / ".cache" / "huggingface" / "hub" / f"models--{model_id.replace('/', '--')}"
    snapshots_dir = cache_root / "snapshots"
    if not snapshots_dir.exists():
        return None

    snapshot_paths = sorted((path for path in snapshots_dir.iterdir() if path.is_dir()), key=lambda item: item.stat().st_mtime, reverse=True)
    return snapshot_paths[0] if snapshot_paths else None


# Loads the tokenizer normally or from cached tokenizer files if needed.
def load_tokenizer_with_fallback(model_id):
    model_source = resolved_model_source if resolved_model_source is not None else model_id
    try:
        return AutoTokenizer.from_pretrained(str(model_source), use_fast=True, local_files_only=resolved_model_source is not None)
    except Exception as exc:
        tokenizer_json_path = Path(model_source) / "tokenizer.json" if resolved_model_source is not None else None
        if tokenizer_json_path is None or not tokenizer_json_path.exists():
            raise exc

        tokenizer_config_path = Path(model_source) / "tokenizer_config.json"
        tokenizer_config = {}
        if tokenizer_config_path.exists():
            tokenizer_config = pd.read_json(tokenizer_config_path, typ="series").to_dict()

        eos_token = extract_special_token_value(tokenizer_config.get("eos_token")) or "<|im_end|>"
        pad_token = extract_special_token_value(tokenizer_config.get("pad_token")) or eos_token
        return AutoTokenizer.from_pretrained(
            str(model_source),
            tokenizer_file=str(tokenizer_json_path),
            eos_token=eos_token,
            pad_token=pad_token,
            use_fast=True,
            local_files_only=True,
        )


resolved_model_source = resolve_cached_model_snapshot(model_name)                  # Prefer an existing local cache to avoid unnecessary downloads.
print(f"resolved_model_source: {resolved_model_source}")

try:
    tokenizer = load_tokenizer_with_fallback(model_name)                           # Load tokenizer with the same fallback idea as Model 2.
    model_source = str(resolved_model_source) if resolved_model_source is not None else model_name
    model = AutoModelForCausalLM.from_pretrained(model_source, local_files_only=resolved_model_source is not None)
except Exception as exc:
    raise RuntimeError(f"Failed to load model or tokenizer '{model_name}': {exc}") from exc

if tokenizer.pad_token is None and tokenizer.eos_token is not None:
    tokenizer.pad_token = tokenizer.eos_token                                      # Reuse EOS when the model does not define a pad token.
if tokenizer.pad_token is None:
    raise RuntimeError("Tokenizer does not define a usable pad_token or eos_token for inference.")
if model.config.pad_token_id is None:
    model.config.pad_token_id = tokenizer.pad_token_id

model = model.to(device)                                                           # Move the model to GPU or CPU once.
model.eval()                                                                       # Disable training-specific behavior.

print(tokenizer)
print(model.__class__.__name__)
```

# 7. Load checkpoint

This stage loads an existing checkpoint and joins it with the current input rows. It also ignores an older API-style fallback answer so those rows can be regenerated locally.

Function definitions:
- `load_checkpoint_frame(path_to_checkpoint)`: Loads the checkpoint, fills missing debug columns when needed, and keeps the newest row per id.
- `save_results(current_results_df)`: Saves both the checkpoint and the ordered final submission file.


```python
def load_checkpoint_frame(path_to_checkpoint):
    if not path_to_checkpoint.exists():
        return pd.DataFrame(columns=checkpoint_columns)

    try:
        checkpoint_df_local = pd.read_csv(path_to_checkpoint, dtype="string")
    except pd.errors.EmptyDataError:
        return pd.DataFrame(columns=checkpoint_columns)
    except pd.errors.ParserError as exc:
        raise ValueError(f"Checkpoint could not be parsed: {exc}") from exc

    for column in checkpoint_columns:
        if column not in checkpoint_df_local.columns:
            checkpoint_df_local[column] = ""                                      # Accept older checkpoint files with fewer debug columns.

    checkpoint_df_local = checkpoint_df_local.loc[:, checkpoint_columns].copy()
    checkpoint_df_local["id"] = checkpoint_df_local["id"].map(normalize_whitespace)
    checkpoint_df_local["answer"] = checkpoint_df_local["answer"].map(normalize_whitespace)
    checkpoint_df_local["status"] = checkpoint_df_local["status"].map(normalize_whitespace)
    checkpoint_df_local["error_message"] = checkpoint_df_local["error_message"].map(normalize_whitespace)
    checkpoint_df_local["attempts"] = checkpoint_df_local["attempts"].fillna("0")
    return checkpoint_df_local.drop_duplicates(subset=["id"], keep="last")


def save_results(current_results_df):
    checkpoint_export_df = current_results_df.loc[:, checkpoint_columns].copy()
    checkpoint_export_df.to_csv(checkpoint_path, index=False)                     # Persist the resumable checkpoint with debug information.

    submission_df = current_results_df.sort_values("input_order").loc[:, ["id", "answer"]].copy()
    submission_df["answer"] = submission_df["answer"].fillna("").astype(str)
    submission_df.to_csv(submission_path, index=False)                            # Persist the final submission in the required format.


checkpoint_df = load_checkpoint_frame(checkpoint_path)
results_df = df.merge(checkpoint_df, on="id", how="left")
results_df["answer"] = results_df["answer"].fillna("")
results_df["status"] = results_df["status"].fillna("pending")
results_df["error_message"] = results_df["error_message"].fillna("")
results_df["attempts"] = pd.to_numeric(results_df["attempts"], errors="coerce").fillna(0).astype(int)

bad_fallback_mask = results_df["answer"].eq("Die Frage konnte im aktuellen Lauf nicht verlaesslich beantwortet werden.")
results_df.loc[bad_fallback_mask, ["answer", "status", "error_message"]] = ["", "pending", "previous API fallback ignored"]

completed_mask = results_df["status"].eq("success") & results_df["answer"].ne("")
print(f"completed_rows: {int(completed_mask.sum())}")
print(f"pending_rows: {int((~completed_mask).sum())}")
display(results_df.head(3))
```

# 8. Batch inference helper

This stage defines the local generation helper. It tokenizes a batch of RAG prompts, runs the model without gradients, removes the prompt part, and returns cleaned answers.

Function definition:
- `generate_batch_answers(prompt_list)`: Generates one cleaned answer for each prompt in a batch.


```python
def generate_batch_answers(prompt_list):
    encoded_input = tokenizer(
        prompt_list,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=max_input_tokens,
    )                                                                              # Tokenize one full batch in a single call.
    encoded_input = {key: value.to(device, non_blocking=pin_memory) for key, value in encoded_input.items()}

    generation_kwargs = {
        "max_new_tokens": max_new_tokens,
        "pad_token_id": tokenizer.pad_token_id,
        "eos_token_id": tokenizer.eos_token_id,
        "do_sample": temperature > 0.0,
    }                                                                              # Build generation settings from the notebook variables.
    if temperature > 0.0:
        generation_kwargs["temperature"] = temperature
        generation_kwargs["top_p"] = top_p

    with torch.inference_mode():
        generated_tokens = model.generate(**encoded_input, **generation_kwargs)     # Run efficient local inference without gradients.

    input_token_count = encoded_input["input_ids"].shape[1]                        # Strip the padded prompt length from every generated sequence.
    answer_list = []
    for output_tokens in generated_tokens:
        answer_tokens = output_tokens[input_token_count:]
        answer_text = tokenizer.decode(answer_tokens, skip_special_tokens=True)
        answer_list.append(normalize_whitespace(answer_text))

    return answer_list
```

# 9. Run RAG inference

This stage runs the full RAG inference loop. For each pending row, it builds a prompt with retrieved RIS context, generates an answer, retries failures, saves checkpoints, and writes the final submission.


```python
pending_index_list = results_df.index[~completed_mask].tolist()                    # Only run inference for rows that are still open.
total_pending = len(pending_index_list)
total_batches = math.ceil(total_pending / batch_size) if total_pending > 0 else 0

if total_pending == 0:
    print("No pending rows found. Existing outputs are reused.")
else:
    start_time = time.time()

    for batch_number in range(total_batches):
        batch_start = batch_number * batch_size
        batch_end = min(batch_start + batch_size, total_pending)
        batch_indices = pending_index_list[batch_start:batch_end]

        prompt_list = []
        for row_index in batch_indices:
            prompt_text = build_prompt(results_df.loc[row_index, "prompt"])
            prompt_list.append(prompt_text)

        batch_answers = None
        last_error_message = ""
        for attempt in range(1, num_retries + 2):
            try:
                batch_answers = generate_batch_answers(prompt_list)
                if len(batch_answers) != len(batch_indices):
                    raise RuntimeError("Generated answer count does not match batch size.")

                for row_index, answer_text in zip(batch_indices, batch_answers):
                    cleaned_answer = normalize_whitespace(answer_text)
                    results_df.at[row_index, "answer"] = cleaned_answer or fallback_answer
                    results_df.at[row_index, "status"] = "success" if cleaned_answer else "fallback"
                    results_df.at[row_index, "error_message"] = ""
                    results_df.at[row_index, "attempts"] = attempt
                break
            except Exception as exc:
                last_error_message = normalize_whitespace(str(exc))
                if attempt <= num_retries:
                    time.sleep(retry_sleep_seconds)

        if batch_answers is None:
            for row_index in batch_indices:
                results_df.at[row_index, "answer"] = fallback_answer
                results_df.at[row_index, "status"] = "error"
                results_df.at[row_index, "error_message"] = last_error_message or "Unknown inference error."
                results_df.at[row_index, "attempts"] = num_retries + 1

        if (batch_number + 1) % save_every_batches == 0 or (batch_number + 1) == total_batches:
            save_results(results_df)

        processed_rows = batch_end
        error_rows = int(results_df["status"].eq("error").sum())
        elapsed_seconds = time.time() - start_time
        print(f"processed {processed_rows}/{total_pending} pending rows | errors: {error_rows} | elapsed: {elapsed_seconds:.1f}s")

    save_results(results_df)

print(f"Saved submission to {submission_path}.")
display(results_df.head(3))
```

This stage checks the final `submission.csv`. It verifies the columns, row count, and id order before showing the status counts and a preview.


```python
if not submission_path.exists():
    raise FileNotFoundError(f"Submission file was not created: {submission_path.resolve()}")

submission_df = pd.read_csv(submission_path, dtype="string")                      # Reload the final submission from disk for validation.

assert list(submission_df.columns) == ["id", "answer"], "Submission must have exactly the columns id and answer."
assert len(submission_df) == len(df), "Submission row count must match the input dataset."
assert submission_df["id"].tolist() == df["id"].tolist(), "Submission ids must preserve the input order."

status_counts = results_df["status"].value_counts(dropna=False)
print(f"Validated {len(submission_df)} submission rows.")
print(status_counts)
display(submission_df.head(10))
```
