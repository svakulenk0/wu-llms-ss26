```python
import pandas as pd
from openai import OpenAI
client = OpenAI(api_key="")

prompt = "Du bist ein österreichischer Steuerrechtsexperte mit der 20-jährigen Erfahrung. Deine Aufgabe ist es, auf die Frage konkret zu antworten. Sage auch die passenden Gesetzesstellen zu dem Rechtsfällen (z.B. § 7 Abs. 1 KStG). Gib Antworten auf Deutsch in ein paar kurzen Sätzen."

orig_df = pd.read_csv("dataset_clean.csv")
```

```python
answers = []

for i, row in orig_df.iterrows():
    prompt = row["prompt"]
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": prompt}],
        max_completion_tokens=250,
        temperature = 0 # we want more focused deterministic output
    )

    answer_text = response.choices[0].message.content.strip()
    answers.append(answer_text)
    print(f"Done: {i+1}/{len(orig_df)} | Preview: {answer_text[:30]}...")
```
