```python
import pandas as pd

# load generated answers
baseline = pd.read_csv("model1_inference (1).csv")
finetune = pd.read_csv("finetune.csv")
rag = pd.read_csv("rag_model_output.csv")
correct = pd.read_csv("Dataset_clean_Answers.csv", sep=';')

# check column names to make sure datasets align correctly before evaluation
baseline.columns, finetune.columns, rag.columns, correct.columns
```

```python
# check answers
df = correct.copy()

df["baseline"] = baseline["answer"]
df["finetune"] = finetune["answer"]
df["rag"] = rag["answer"]

df.head()
```

```python
# save merged dataset containing model outputs and reference answers for later evaluation
df.to_csv("evaluation_dataset.csv", index=False)
```

```python
from bert_score import score

# calculate bertscore precision, recall, and f1 between baseline model answers and reference answers
P_base, R_base, F_base = score(
    df["baseline"].astype(str).tolist(),        # baseline model prediction
    df["correct_answer"].astype(str).tolist(),  # ground truth reference answers
    lang="de"                                   # german setting for embeddings
)

# compute bertscore for fine-tuned mistral model answers
P_ft, R_ft, F_ft = score(
    df["finetune"].astype(str).tolist(),        # fine-tuned model predictions
    df["correct_answer"].astype(str).tolist(),  # ground truth reference answers
    lang="de"
)

# compute bertscore for rag model answers
P_rag, R_rag, F_rag = score(
    df["rag"].astype(str).tolist(),             # rag model predictions
    df["correct_answer"].astype(str).tolist(),  # ground truth reference answers
    lang="de"
)
```

```python
# compute average bertscore f1 value across all questions
baseline_score = F_base.mean().item()
finetune_score = F_ft.mean().item()
rag_score = F_rag.mean().item()

baseline_score, finetune_score, rag_score
```

```python
# create df summarizing bertscore f1 results for all evaluated models
results = pd.DataFrame({
    "Model": ["Baseline", "Fine-tuned", "RAG"],
    "BERTScore_F1": [
        # average semantic similarity score
        baseline_score,
        finetune_score,
        rag_score
    ]
})

results
```

```python
# save final comparison table with bertscore results for all models to csv
results.to_csv("model_comparison.csv", index=False)
```

```python
# identify model with highest bertscore f1 value
best_model = results.loc[results["BERTScore_F1"].idxmax()]
best_model
```

```python
# attach bertscore f1 values for each individual question to the dataframe
# semantic similarity scores
df["baseline_score"] = F_base.tolist()
df["finetune_score"] = F_ft.tolist()
df["rag_score"] = F_rag.tolist()

# save extended evaluation dataset including similarity scores
df.to_csv("evaluation_with_scores.csv", index=False)
```

```python
from rouge_score import rouge_scorer
import pandas as pd
```

```python
# replace missing values with empty strings to avoid errors during computation
df = df.fillna("")

# remove placeholder error outputs from model predictions before evaluation
df["baseline"] = df["baseline"].replace("ERROR", "")
df["finetune"] = df["finetune"].replace("ERROR", "")
df["rag"] = df["rag"].replace("ERROR", "")
```

```python
# initialize rouge scorer with unigram, bigram, and longest common subsequence overlap metrics
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# function to compute average rouge scores across all predictions
def compute_rouge(predictions, references):
    scores = {"rouge1": [], "rouge2": [], "rougeL": []}  # store per-example rouge scores

    # iterate through answers and reference answers
    for pred, ref in zip(predictions, references):
        score = scorer.score(ref, pred)  # compute rouge overlap between prediction and ground
        scores["rouge1"].append(score["rouge1"].fmeasure)  # store unigram score
        scores["rouge2"].append(score["rouge2"].fmeasure)  # store bigram score
        scores["rougeL"].append(score["rougeL"].fmeasure)  # store longest sequence score

    # return average rouge scores across whole dataset
    return {metric: sum(vals)/len(vals) for metric, vals in scores.items()}
```

```python
# compute scores for baseline model predictions compared to reference answers
rouge_baseline = compute_rouge(df["baseline"], df["correct_answer"])

# compute fine-tuned (mistral) model predictions
rouge_finetune = compute_rouge(df["finetune"], df["correct_answer"])

# compute rag model predictions
rouge_rag = compute_rouge(df["rag"], df["correct_answer"])

rouge_baseline, rouge_finetune, rouge_rag
```

```python
# create df summarizing rouge results
rouge_results = pd.DataFrame({
    "Model": ["Baseline", "Fine-tuned", "RAG"],  # names of evaluated models

    "ROUGE-1": [
        rouge_baseline["rouge1"],   # unigram overlap between predictions and reference
        rouge_finetune["rouge1"],   # for fine-tuned model
        rouge_rag["rouge1"]         # for rag model
    ],

    "ROUGE-2": [
        rouge_baseline["rouge2"],   # bigram overlap between predictions and reference
        rouge_finetune["rouge2"],   # for fine-tuned model
        rouge_rag["rouge2"]         # for rag model
    ],

    "ROUGE-L": [
        rouge_baseline["rougeL"],   # longest common subsequence similarity: baseline model
        rouge_finetune["rougeL"],   # fine-tuned model
        rouge_rag["rougeL"]         # rag model
    ]
})

# display rounded results
rouge_results.round(4)
```

```python
# save rouge evaluation table for all models to csv
rouge_results.to_csv("rouge_results.csv", index=False)
```
