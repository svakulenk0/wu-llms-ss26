```python
from groq import Groq
import pandas as pd
from tqdm import tqdm

# API Key setzen
client = Groq(api_key="API_KEY")

# Daten laden
df = pd.read_csv("dataset_clean.csv")

print("Daten geladen:", len(df), "Fragen")

results = []

print("Starte Inference (komplett)...")

for question in tqdm(df["prompt"]):

    try:
        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "system",
                    "content": (
                        "Du bist ein Experte für österreichisches Steuerrecht. "
                        "Antworte präzise, fachlich korrekt und in maximal 2-3 Sätzen. "
                        "Keine Einleitung. Keine Wiederholung der Frage."
                    )
                },
                {
                    "role": "user",
                    "content": question
                }
            ],
            temperature=0.1
        )

        answer = response.choices[0].message.content.strip()

    except Exception as e:
        print("Fehler:", e)
        answer = "ERROR"

    results.append({
        "question": question,
        "answer": answer
    })

# speichern
output_file = "results_model1_groq_full.csv"
pd.DataFrame(results).to_csv(output_file, index=False)

print("Fertig gespeichert:", output_file)
```
