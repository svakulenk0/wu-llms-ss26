```python
from transformers import pipeline, BitsAndBytesConfig
import torch

quantization_config = BitsAndBytesConfig(load_in_4bit=True)

pipe = pipeline(
    "text-generation",
    model="meta-llama/Llama-3.1-8B-Instruct",
    torch_dtype=torch.float16,
    device_map="auto",
    model_kwargs={"quantization_config": quantization_config}
)
```

```python
print("Model loaded.")
```

```python
system_message = {
    "role": "system",
    "content": (
        "Du beantwortest Fragen zum österreichischen Steuerrecht kurz und sachlich. "
        "Erfinde keine Rechtsgrundlagen. Wenn du unsicher bist, sage das kurz."
    ),
}

def get_answer(user_text):
    messages = [
        system_message,
        {"role": "user", "content": user_text},
    ]

    output = pipe(messages, max_new_tokens=150, do_sample=False)
    return output[0]["generated_text"][-1]["content"].strip()
```

```python
answer_file_path = "/content/drive/MyDrive/Austrian Tax Law LLM Prompt Assignment.xlsx"
answers_df = pd.read_excel(answer_file_path, sheet_name="Dataset")
answers_df = answers_df[["id", "correct_answer", "sources"]].dropna(subset=["id"])
```

```python
full_df = prompt_df.reset_index(drop=True).copy()

baseline_answers = []

for i, row in full_df.iterrows():
    answer = get_answer(row["prompt"])
    baseline_answers.append(answer)
    print(f"{i + 1}/{len(full_df)} done")

full_df["baseline_answer"] = baseline_answers
```

```python
results_df = full_df.merge(answers_df, on="id", how="left")
```

```python
results_df.to_csv("/content/baseline_results.csv", index=False)
```
