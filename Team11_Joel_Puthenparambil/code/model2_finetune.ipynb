## Generate training data from RIS statute texts

We scrape Austrian law §§ from RIS and generate Q&A training pairs using templates.
No API needed — each answer is directly grounded in a real statute section.

## Cell 3 — Scrape Statute Text from RIS

This cell fetches the full text of five Austrian laws from `ris.bka.gv.at` and splits each into individual § sections.

- `requests.get(url, timeout=60)`: downloads the HTML page for each law by its `Gesetzesnummer`.
- `BeautifulSoup(...).get_text(separator="\n", strip=True)`: extracts plain text from the HTML.
- `re.split(r'(?=§\s*\d+[a-z]?\.)', text)`: splits the text at every § boundary.
- `part[:1500]`: caps each chunk at 1 500 characters to keep training examples concise.
- `chunks.append({...})`: stores the law name, section number, and text for each §.

**Project Impact**

Produces the raw legal text used to build training examples. More sections scraped = more diverse training pairs and better coverage of the Austrian tax law topics in the test set.

```python
# Cell 3 — Scrape laws from RIS (same approach as Model 3)

LAWS = {
    "EStG 1988":   "10004570",
    "KStG 1988":   "10004569",
    "UStG 1994":   "10004873",
    "GrEStG 1987": "10004531",
    "BAO":         "10003940",
}

def scrape_law(name, gesetzesnummer):
    """Fetch full text of a law from RIS and split into § chunks."""
    url = f"https://www.ris.bka.gv.at/GeltendeFassung.wxe?Abfrage=Bundesnormen&Gesetzesnummer={gesetzesnummer}"
    print(f"  Fetching {name}...")
    resp = requests.get(url, timeout=60)
    resp.encoding = "utf-8"
    soup = BeautifulSoup(resp.text, "html.parser")
    text = soup.get_text(separator="\n", strip=True)

    # split on § boundaries
    parts = re.split(r'(?=§\s*\d+[a-z]?\.)', text)
    chunks = []
    for part in parts:
        part = part.strip()
        if not part or not part.startswith("§"):
            continue
        m = re.match(r'(§\s*\d+[a-z]?)\.?', part)
        section = m.group(1) if m else "?"
        if len(part) < 80:  # skip tiny fragments
            continue
        chunks.append({"law": name, "section": section, "text": part[:1500]})
    print(f"  → {name}: {len(chunks)} sections")
    return chunks

# scrape all laws
all_sections = []
for name, gnr in LAWS.items():
    try:
        all_sections.extend(scrape_law(name, gnr))
    except Exception as e:
        print(f"  ERROR: {name}: {e}")

print(f"\nTotal: {len(all_sections)} statute sections scraped.")
```

## Cell 4 — Generate Q&A Training Pairs

This cell creates template-based question–answer pairs from the scraped statute sections.

- `TEMPLATES`: four question/answer template pairs with varying phrasing (e.g. "Was regelt …?", "Was besagt …?").
- `make_summary(text)`: extracts the first ~300 characters of meaningful content from a § section.
- `TEMPLATES[len(training_pairs) % len(TEMPLATES)]`: rotates through the four templates so each section gets a different question style.
- `training_pairs.append({...})`: stores each generated question and answer.

**Project Impact**

Produces ~750 training pairs without any external API. The model learns the expected answer style and citation format. Because examples are template-based, the model learns structure rather than deep legal knowledge.

```python
# Cell 4 — Generate Q&A training pairs from statute sections
# Template-based: each pair is grounded in an actual § text

# question templates — varied to teach the model different answer styles
TEMPLATES = [
    ("Was regelt {section} {law}?",
     "{section} {law} regelt Folgendes: {summary}"),
    ("Was besagt {section} {law}?",
     "Gemäß {section} {law}: {summary}"),
    ("Welche Regelung enthält {section} {law}?",
     "Die Regelung in {section} {law} besagt: {summary}"),
    ("Was ist der Inhalt von {section} {law}?",
     "Der Inhalt von {section} {law} umfasst: {summary}"),
]

def make_summary(text, max_len=300):
    """Extract the first meaningful sentences from a statute section."""
    # remove the § header line
    lines = text.split("\n")
    content = " ".join(line.strip() for line in lines[1:] if line.strip())
    # take first ~300 chars ending at a sentence boundary
    if len(content) > max_len:
        cut = content[:max_len].rfind(".")
        if cut > 100:
            content = content[:cut+1]
        else:
            content = content[:max_len] + "..."
    return content

# generate pairs
training_pairs = []
for sec in all_sections:
    summary = make_summary(sec["text"])
    if len(summary) < 30:  # skip sections with too little content
        continue

    # pick one template per section (rotate through templates)
    tmpl_q, tmpl_a = TEMPLATES[len(training_pairs) % len(TEMPLATES)]
    q = tmpl_q.format(section=sec["section"], law=sec["law"])
    a = tmpl_a.format(section=sec["section"], law=sec["law"], summary=summary)
    training_pairs.append({"question": q, "answer": a})

print(f"Generated {len(training_pairs)} training pairs.")
# show a few examples
for p in training_pairs[:3]:
    print(f"\nQ: {p['question']}")
    print(f"A: {p['answer'][:150]}...")
```

## Fine-tune with QLoRA

## Cell 5 — Load Base Model (fp16, No Quantization)

This cell downloads `TinyLlama-1.1B-Chat-v1.0` and loads it in half-precision.

- `AutoTokenizer.from_pretrained(MODEL_NAME)`: downloads the matching tokenizer.
- `tokenizer.pad_token = tokenizer.eos_token`: sets the padding token (TinyLlama has none by default).
- `tokenizer.padding_side = "right"`: pads sequences on the right, as required for causal language models.
- `AutoModelForCausalLM.from_pretrained(..., torch_dtype=torch.float16, device_map="auto")`: loads the 1.1B model in fp16 onto the GPU automatically.

**Project Impact**

TinyLlama at 1.1B parameters fits in GPU memory without quantization. Choosing this model avoids bitsandbytes compatibility issues (requires CUDA sm_70+) while still training efficiently on a free GPU.

```python
# Cell 5 — Load TinyLlama in fp16 (no bitsandbytes — P100 CUDA sm_60 incompatible)
# TinyLlama 1.1B fits easily in 16GB VRAM, trains fast on P100

MODEL_NAME = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# load in fp16 — no quantization needed at 1.1B
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto"
)

print(f"Model loaded: {MODEL_NAME}")
print(f"Parameters: {sum(p.numel() for p in model.parameters()) / 1e9:.1f}B")
```

## Cell 6 — Apply LoRA Configuration

This cell wraps the base model with lightweight LoRA adapters.

- `LoraConfig(r=16, lora_alpha=32, target_modules=["q_proj", "v_proj"])`: defines the adapter rank, scaling factor, and which attention layers to target.
- `lora_dropout=0.05`: applies dropout to the adapters to reduce overfitting on the small training set.
- `get_peft_model(model, lora_config)`: injects the LoRA layers into the model.
- `model.print_trainable_parameters()`: prints what fraction of weights will be updated (~0.2%).

**Project Impact**

LoRA makes fine-tuning feasible on a free GPU — only ~2 M adapter weights are trained instead of all 1.1B base weights, drastically reducing memory usage and training time.

```python
# Cell 6 — Apply LoRA config (no prepare_model_for_kbit_training — not using 4-bit)
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()  # should be ~0.5% of total
```

## Cell 7 — Format Training Data as Chat Template

This cell converts each Q&A pair into the format SFTTrainer expects.

- `tokenizer.apply_chat_template(messages, tokenize=False)`: renders the system prompt + question + answer into TinyLlama's chat template string.
- `{"role": "user", ...}` and `{"role": "assistant", ...}`: defines the two turns of the conversation.
- `Dataset.from_list(training_pairs).map(format_chat)`: creates a HuggingFace Dataset and applies the formatting to every example.

**Project Impact**

Correct formatting is critical — if the chat template is wrong, the model trains on malformed sequences and produces incoherent output at inference time.

```python
# Cell 7 — Format training data as instruction-tuning chat

def format_chat(pair):
    """Format a Q&A pair as a Mistral chat conversation."""
    messages = [
        {"role": "user", "content": SYSTEM_PROMPT + "\n\n" + pair["question"]},
        {"role": "assistant", "content": pair["answer"]}
    ]
    text = tokenizer.apply_chat_template(messages, tokenize=False)
    return {"text": text}

# create HuggingFace dataset
train_data = Dataset.from_list(training_pairs)
train_data = train_data.map(format_chat)

print(f"Training dataset: {len(train_data)} examples")
print(f"\nExample formatted text (first 300 chars):")
print(train_data[0]["text"][:300])
```

## Cell 8 — Fine-Tune with SFTTrainer

This cell runs the actual LoRA fine-tuning on the generated training pairs.

- `SFTConfig(num_train_epochs=3, per_device_train_batch_size=4, learning_rate=2e-4, fp16=True)`: sets the core training hyperparameters.
- `gradient_accumulation_steps=2`: simulates a larger effective batch size (8) to stabilise training.
- `max_grad_norm=0.3`: clips gradients to prevent unstable updates.
- `SFTTrainer(model, train_dataset, args, processing_class=tokenizer)`: sets up the trainer with the model and data.
- `trainer.train()`: runs 3 epochs over all training examples.
- `model.save_pretrained("./lora_adapter")`: saves the trained LoRA adapter weights to disk.

**Project Impact**

This is the core training cell. After it completes the model has been adapted to produce German legal answers in the expected citation style. Training takes roughly 10 minutes on a free T4 GPU.

```python
# Cell 8 — Train with SFTTrainer
from trl import SFTTrainer, SFTConfig

training_args = SFTConfig(
    output_dir="./lora_output",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=2,
    learning_rate=2e-4,
    warmup_steps=20,
    weight_decay=0.01,
    logging_steps=10,
    save_strategy="epoch",
    fp16=True,
    max_grad_norm=0.3,
    report_to="none",
)

trainer = SFTTrainer(
    model=model,
    train_dataset=train_data,
    args=training_args,
    processing_class=tokenizer,
)

print("Starting training...")
trainer.train()
print("Training done!")

model.save_pretrained("./lora_adapter")
print("LoRA adapter saved.")
```

## Cell 10 — Run Inference on All 643 Questions

This cell generates an answer for every question using the fine-tuned model.

- `model.gradient_checkpointing_disable()` and `model.config.use_cache = True`: re-enables the KV cache for faster generation (disabled during training).
- `tokenizer.apply_chat_template(messages, add_generation_prompt=True)`: formats each question as a chat prompt.
- `model.generate(..., do_sample=True, temperature=0.5, top_p=0.9, repetition_penalty=1.3, max_new_tokens=150)`: generates up to 150 tokens with controlled sampling.
- `eos_token_id=[tokenizer.eos_token_id, user_token_id]`: stops generation when the model would start a second user turn.
- `output[0][inputs["input_ids"].shape[1]:]`: decodes only the newly generated tokens, not the prompt.

**Project Impact**

Produces all 643 answers. The repetition penalty prevents looping, and the dual EOS token stops generation cleanly before a second conversation turn begins.

```python
# Cell 10 — Inference auf allen 643 Fragen
results = []  # immer neu starten

# modell für inference vorbereiten
model.gradient_checkpointing_disable()
model.config.use_cache = True
model.eval()

user_token_id = tokenizer.convert_tokens_to_ids("<|user|>")

for i, row in df.iterrows():
    try:
        messages = [{"role": "user", "content": SYSTEM_PROMPT + "\n\n" + row["prompt"]}]
        prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        with torch.no_grad():
            output = model.generate(
                **inputs,
                max_new_tokens=150,
                do_sample=True,
                temperature=0.5,
                top_p=0.9,
                repetition_penalty=1.3,
                pad_token_id=tokenizer.eos_token_id,
                eos_token_id=[tokenizer.eos_token_id, user_token_id],
            )

        new_tokens = output[0][inputs["input_ids"].shape[1]:]
        answer = tokenizer.decode(new_tokens, skip_special_tokens=True).strip()
        if not answer:
            answer = "Keine Antwort generiert."
        results.append({"id": row["id"], "answer": answer})

    except Exception as e:
        print(f"Fehler bei {row['id']}: {e}")
        results.append({"id": row["id"], "answer": "Fehler bei der Generierung."})

    if len(results) % 50 == 0:
        print(f"  Progress: {len(results)}/{len(df)}")

print(f"Done. {len(results)} Antworten generiert.")
```

- `df[["id"]].merge(result_df, on="id", how="left")`: reorders rows to match the original dataset, filling any missing answers with NaN.

```python
# Cell 11 — Save results to CSV
output_path = "model2_finetuned.csv"
result_df = pd.DataFrame(results)
print(f"Results: {len(result_df)} Zeilen")

if len(result_df) == 0:
    print("ERROR: results ist leer — Cell 10 nochmal ausführen!")
else:
    # merge statt loc (robuster bei ID-Abweichungen)
    final_df = df[["id"]].merge(result_df, on="id", how="left")
    final_df.to_csv(output_path, index=False)
    print(f"Gespeichert: {output_path}")
    print("Download: File Browser links → Rechtsklick → Download")
```

## Cell 12 — Validate Submission Format

This cell checks the output CSV for the required submission format.

- `assert list(sub.columns) == ["id", "answer"]`: verifies the column names are exactly correct.
- `assert len(sub) == len(ref)`: confirms all 643 rows are present.
- `assert sub["id"].tolist() == ref["id"].tolist()`: checks that ID order matches the original dataset.
- `assert sub["answer"].notna().all()`: ensures no answer is missing (NaN).
- `assert (sub["answer"].astype(str).str.strip() != "").all()`: ensures no answer is blank.
- `assert top_freq < 20`: catches repeated identical answers indicating a crash loop.

**Project Impact**

A final safety check before submission — catches formatting errors that would silently break automated evaluation and cost points.

```python
# Cell 12 — Validate submission format
ref = pd.read_csv(csv_path)
sub = pd.read_csv(output_path)

assert list(sub.columns) == ["id", "answer"], "Wrong columns"
assert len(sub) == len(ref), f"Expected {len(ref)} rows, got {len(sub)}"
assert sub["id"].tolist() == ref["id"].tolist(), "IDs don't match"
assert sub["id"].nunique() == len(sub), "Duplicate IDs"
assert sub["answer"].notna().all(), "Null answers found"
assert (sub["answer"].astype(str).str.strip() != "").all(), "Blank answers found"
top_freq = sub["answer"].value_counts().iloc[0]
assert top_freq < 20, f"Repeated answer {top_freq}x — possible crash"

print(f"OK — {output_path} is valid ({len(sub)} rows)")
print("\nSample answers:")
for _, row in sub.head(3).iterrows():
    print(f"\n{row['id']}: {row['answer'][:150]}...")
```
