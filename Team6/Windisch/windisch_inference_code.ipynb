```python
import torch                            # used to detect if a GPU is available
import pandas as pd                     
from transformers import pipeline       # Hugging Face pipeline for running AI models

# checking if GPU is available
device_id = 0 if torch.cuda.is_available() else -1           # 0 = GPU, -1 = CPU only
if device_id == 0:
    print("GPU (CUDA) gefunden!")                            # GPU found, model will run fast
else:
    print("WARNUNG: Keine GPU gefunden. LÃ¤uft auf CPU.")     # No GPU, model will run slow on CPU
```

```python
# loading the dataset
df = pd.read_csv("dataset_clean.csv")                # dataset with promts
print("Spalten im Dataset:", df.columns.tolist())    # printing columns to look at structure
pd.head()                                            
```

```python
# adding a new column "correct answer" which will later hold the models outputs
df["correct_answer"] = "nix" # for now we put a placeholder "nix"
```

```python
# storing the column names as variables
COL_ID     = "id"
COL_PROMPT = "prompt"
COL_ANSWER = "correct_answer"
```

```python
# loading the models
# loeading gpt-2 large as a text generator pipeline
generator = pipeline('text-generation', model='gpt2-large', device=device_id)   # device=device_id sends the model to GPU/CPU

# setting the padding token as the end of sequence token
# GPT-2 has no pad token so we use end of sequence to prevent tokenizer warning
generator.tokenizer.pad_token_id = generator.tokenizer.eos_token_id

print("Pipeline loaded.") # confirmation if model loaded successfully
```

```python
# running the inference loop
results = [] # empty list that will collect one result dictionary per promt

for i, row in df.iterrows(): # looping though all rows
    print(f"Processing {i+1}/{len(df)}: {row[COL_ID]}") # showing what id is currently processed

    # sending the promt to the model and generating a response
    output = generator(
        row[COL_PROMPT],         # the question / promt text that is being processed
        max_new_tokens=100,      # generating at most 100 tokens
        num_return_sequences=1,  # returning only 1 answer per promt
        truncation=True,         # truncating (cutting off) end of input if it exceeds maximum model length
        return_full_text=False   # returning only the generated anser, not the origninal prompt again
    )

    # saving the results for the row as a dictionary and adding it to the results list
    results.append({
        "id":             row[COL_ID],                                            # original prompt id
        "prompt":         row[COL_PROMPT],                                        # original promt text
        "model_answer":   output[0]['generated_text'].replace('\n', ' ').strip()  # generated output, removing newliens and trimming whitespaces
    })

print("Done.") # all promts have been processed
```

```python
# creating a output dataframe
results_df = pd.DataFrame(results)                              # converting the list of results into a df
results_df.to_csv("results_model1_inference.csv", index=False)  # saving as csv
print("Saved.")
results_df.head()
```
