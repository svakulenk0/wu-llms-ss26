```python
import requests                                    # HTTP library for making API calls
from bs4 import BeautifulSoup                      # HTML parser
import os                                          # For creating directories
import time                                        # For adding delays between requests

os.makedirs("raw_laws", exist_ok=True)             # Create the "raw_laws" folder; do nothing if it already exists

# -------------------------------------------------------
# Die Gesetzesnummern auf RIS (Bundesrechtsnummern):
# EStG 1988  = 10004570
# UStG 1994  = 10004873
# KStG 1988  = 10005025
# GrEStG     = 10002382
# -------------------------------------------------------

gesetze = {                                        # Dictionary mapping law short names to their official RIS law numbers
    "EStG": "10004570",                            # Einkommensteuergesetz 1988
    "UStG": "10004873",                            # Umsatzsteuergesetz 1994
    "KStG": "10005025",                            # Körperschaftsteuergesetz 1988
    "GrEStG": "10002382"                           # Grunderwerbsteuergesetz 1987
}

def download_gesetz(name, gesetzesnummer):          # Define a function to download one law by name and RIS number
    """
    Lädt ein Gesetz von der RIS API herunter.
    RIS liefert eine Liste aller Paragraphen als JSON zurück.
    """
    
    print(f"Lade {name}...")                        # Print which law is currently being downloaded
    
    # RIS API Endpoint
    url = "https://data.bka.gv.at/ris/api/v2.6/Bundesrecht"  # Base URL for the Austrian RIS federal law API
    
    params = {                                     # Query parameters sent along with the API request
        "Applikation": "BrKons",                   # "BrKons" = consolidated (Konsolidierte) federal law version
        "Gesetzesnummer": gesetzesnummer,           # Identifies which specific law to retrieve
        "Fassung.FassungVom": "2025-01-01"         # Retrieve the version of the law valid as of January 1, 2025
    }
    
    response = requests.get(url, params=params)    # Send the GET request to the RIS API with the above parameters
    
    # Checking HTTP Status
    print(f"  Status: {response.status_code}")     # Print the HTTP response code (200 = success)
    
    if response.status_code == 200:                # Only save the file if the request was successful
        # Rohantwort als Textdatei speichern zur Kontrolle
        with open(f"raw_laws/{name}_raw.json", "w", encoding="utf-8") as f:  # Open a new file for writing with UTF-8 encoding
            f.write(response.text)                 # Write the raw JSON response text to disk
        print(f"  ✓ Gespeichert als raw_laws/{name}_raw.json")  # Confirm successful save
    else:
        print(f"  ✗ Fehler beim Download")         # Print error message if the request failed
    
    time.sleep(1)                                  # Wait 1 second between requests to avoid overloading the server

# Downloading all law texts
for name, nummer in gesetze.
```
