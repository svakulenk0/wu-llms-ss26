```python
import json                                        # Standard library for JSON handling

with open("raw_laws/EStG_raw.json", "r", encoding="utf-8") as f:  # Open the EStG raw JSON file
    data = json.load(f)                            # Parse JSON into a Python object

erster_eintrag = data['OgdSearchResult']['OgdDocumentResults']['OgdDocumentReference'][0]
                                                   # Navigate to the first paragraph entry

# Dokumentliste anschauen
print("=== Dokumentliste Keys ===")               # Section header
dokumentliste = erster_eintrag['Data']['Dokumentliste']  # Access the document list section of the entry
print(type(dokumentliste))                         # Print whether Dokumentliste is a dict or list
print(list(dokumentliste.keys()))                  # Print all keys inside Dokumentliste

# Eine Ebene tiefer
print("\n=== Inhalt von Dokumentliste ===")        # Section header
for key in dokumentliste.keys():                   # Iterate over every key inside Dokumentliste
    inhalt = dokumentliste[key]                    # Get the value for each key
    print(f"\nKey '{key}' (Typ: {type(inhalt)}):")  # Print the key name and its data type
    print(str(inhalt)[:300])                       # Print the first 300 characters of the value

# Allgemeine Metadaten anschauen (könnte Paragraph-Nummer etc. enthalten)
print("\n=== Allgemeine Metadaten ===")            # Section header
print(erster_eintrag['Data']['Metadaten']['Allgemein'])  # Print general metadata (title, date, etc.)

# Bundesrecht Metadaten (könnte § Nummer enthalten)
print("\n=== Bundesrecht Metadaten ===")           # Section header
print(erster_eintrag['Data']['Metadaten']['Bundesrecht'])  # Print federal law specific metadata (paragraph number, abbreviation, etc.)
```
